```python
import pprint
pp = pprint.PrettyPrinter(indent=4)
import matplotlib.pyplot as plt

import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
diabetes = datasets.load_diabetes()
diabetes.feature_names
#pp.pprint(diabetes)
```

```python
diabetes_X = diabetes.data[:, np.newaxis, 2] # Change 2 if you want to choose different feature col 
#pp.pprint(diabetes_X)
```

```python
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test  = diabetes_X[-20:]
#pp.pprint(diabetes_X_train)
```

```python
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test  = diabetes.target[-20:]
#pp.pprint(diabetes_y_train)
```

```python
# Plot train data
plt.plot(diabetes_X_train, diabetes_y_train, 'o')
plt.show()
```

```python
regr = linear_model.LinearRegression()
regr.fit(diabetes_X_train, diabetes_y_train)
diabetes_y_pred = regr.predict(diabetes_X_test)
```

```python
print('Coefficients: \n', regr.coef_)
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(diabetes_y_test, diabetes_y_pred))
```

```python
# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test,  color='black')
plt.plot(diabetes_X_test, diabetes_y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
