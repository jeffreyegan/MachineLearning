```python
import numpy as np
from sklearn.datasets.samples_generator import make_regression
import pylab
from scipy import stats
```

```python
def gradient_descent_2(alpha, x, y, numIterations):
    m = x.shape[0] # number of samples
    theta = np.ones(2)
    costarr = [None] * numIterations
    x_transpose = x.transpose()
    for iter in range(0, numIterations):
        hypothesis = np.dot(x, theta)
        loss = hypothesis - y
        J = np.sum(loss ** 2) / (2 * m)  # cost
        costarr[iter] = J
        #print("iter %s | J: %.3f" % (iter, J))
        gradient = np.dot(x_transpose, loss) / m
        theta = theta - alpha * gradient  # update
    return [theta,costarr]
```

```python
NUMBER_SAMPLES = 100

x, y = make_regression(n_samples=NUMBER_SAMPLES, n_features=1, n_informative=1,
                           random_state=0, noise=35)
pylab.plot(x,y, 'o')
pylab.ylabel('y')
pylab.xlabel('x')
pylab.title('Randomly Generated Dataset')
pylab.show()
```

```python
NUM_ITERATIONS = 1000
LEARNING_RATE = 0.01

m, n = np.shape(x)
x_ = np.c_[ np.ones(m), x] # insert column
#change the learning rate to see what happens
alpha = LEARNING_RATE # learning rate
[theta,costarr] = gradient_descent_2(alpha, x_, y, numIterations=NUM_ITERATIONS) # try messing around with numIterations! 
```

```python
pylab.plot(costarr)
pylab.xlabel('Iterations')
pylab.ylabel('Cost')
pylab.show()
```

```python
# plot
for i in range(x_.shape[1]):
        y_predict = theta[0] + theta[1]*x_
pylab.plot(x_[:,1],y,'o')
pylab.plot(x_,y_predict,'k-')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.show()
print("Done!")
```
